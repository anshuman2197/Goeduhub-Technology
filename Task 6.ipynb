# 1 Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed represents a car’s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.
         Days=[1,2,3,4,5,6,7,8]

         Speed=[60,62,61,58,56,57,46,63]​​​​​

```python
import matplotlib.pyplot as plt
Days=[1,2,3,4,5,6,7,8]
Speed=[60,62,61,58,56,57,46,63]
plt.plot(Days,Speed)
plt.title("Car Speed Measurement")
plt.xlabel("Days")
plt.ylabel("Speed")
plt.show()
```

# 4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
X=np.arange(0,math.pi*2,0.05)
Y1 = np.sin(X)
Y2 = np.cos(X)
Y3 = np.tan(X)
Y4 = np.tanh(X)
figure, axis = plt.subplots(2, 2)
axis[0, 0].plot(X, Y1)
axis[0, 0].set_title("Sine Function")
axis[0, 1].plot(X, Y2)
axis[0, 1].set_title("Cosine Function")
axis[1, 0].plot(X, Y3)
axis[1, 0].set_title("Tangent Function")
axis[1, 1].plot(X, Y4)
axis[1, 1].set_title("Tanh Function")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
ti = pd.read_csv("C://Users//Mypc//Desktop//Goeduhub Technology//titanic.csv",skip_blank_lines=True)
ti
```

```python
count=ti['sex'].value_counts()
colors = ['blue', 'orange']
f, (ax1, ax2) = plt.subplots(1, 2)
ax1.pie(count.values.tolist(), labels=count.index.values.tolist(), colors=colors,startangle=90,
        autopct='%.1f%%')
ax1.set_title('Male/Female proportion')
```

```python
colors = {'male':'blue', 'female':'orange'}
ax2.scatter(ti.fare,ti.age,ti['sex'].iloc[:-1].apply(lambda x:colors[x]))
ax2.set_xlabel('Fare')
ax2.set_ylabel('Age')
ax2.set_title('Fare paid and the Age')
```
